```python
import pandas, numpy
import matplotlib, matplotlib.pyplot

matplotlib.rcParams.update({'font.size':18,'font.family':'Arial','xtick.labelsize':14,'ytick.labelsize':14})
matplotlib.rcParams['pdf.fonttype']=42
```

```python
regulon_file = '/Users/alomana/cdi/miner_results_merged/subtype_analysis_results/regulons_activity_heatmap.csv'
```

```python
df = pandas.read_csv(regulon_file, sep=',', index_col=0)
print(df.shape)
df.head()
```

```python
# Regulon activity heatmap
matplotlib.pyplot.figure(None, (8,8))
matplotlib.pyplot.imshow(df, cmap="bwr", vmin=-1, vmax=1, aspect=0.275)
matplotlib.pyplot.grid(False)
```

```python
all_columns = df.columns.to_list()
selected_columns = [element for element in all_columns if 'cdi' in element]
print(selected_columns)
```

```python
selected = df.loc[:, selected_columns]
selected.head()
```

```python
matplotlib.pyplot.figure(None, (8,8))
matplotlib.pyplot.imshow(selected, cmap="bwr", vmin=-1, vmax=1, aspect=0.275)
matplotlib.pyplot.grid(False)
```
