```python
import pandas, os
```

# 0. user defined variables

```python
mmrf_file = '/Users/alomana/cdi/data/MMRF_CoMMpass_IA14a_E74GTF_Salmon_Gene_TPM.txt'
cdi_folder = '/Volumes/omics4tb2/alomana/projects/cdi/results/salmon/'
merged_output_file = '/Users/alomana/cdi/merge_results/mmrf14a_cdi.txt'
```

# 1. read MMRF file as a pandas dataframe

```python
mmrf = pandas.read_csv(mmrf_file, sep='\t', index_col=0)
print(mmrf.shape)
mmrf.head()
```

# 2. read CDI samples as a pandas dataframe

```python
found = os.listdir(cdi_folder)
samples = [element for element in found if '_R' in element] 
print(samples)
```

```python
for sample in samples[:1]:
    label = 'cdi_' + sample[:2]
    print(label)
    working_file = cdi_folder + sample + '/{}_salmon_e74GTF_filtered_genes.sf'.format(sample)
    cdi = pandas.read_csv(working_file, sep='\t', index_col=0)
    cdi.rename(columns={"TPM": label}, inplace=True)
    print(cdi.shape)
    #print(cdi.head())
    
for sample in samples[1:]:
    label = 'cdi_' + sample[:2]
    print(label)
    working_file = cdi_folder + sample + '/{}_salmon_e74GTF_filtered_genes.sf'.format(sample)
    df = pandas.read_csv(working_file, sep='\t', index_col=0)
    cdi[label] = df['TPM']
    print(cdi.shape)
    #print(cdi.head())
    
print('final df')
print(cdi.shape)
cdi.head()
```

# 3. data intersection

```python
# check intersection numbers
mmrf_genes = mmrf.index.to_list()
cdi_genes = cdi.index.to_list()
print(len(mmrf_genes), len(cdi_genes))
intersect = list(set(mmrf_genes) & set(cdi_genes))
print(len(intersect))
```

```python
merged = mmrf.merge(cdi, right_on='Name', how='inner', left_index=True)
print(merged.shape)
merged.head()
```

# 4. store file

```python
merged.to_csv(path_or_buf=merged_output_file, sep='\t')
```
